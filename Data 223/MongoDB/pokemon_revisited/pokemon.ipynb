```python
def get_pokemon_details(id):
    response= requests.get(f'https://pokeapi.co/api/v2/pokemon/{id}') 
    pokemon_data=response.json() 
    id=pokemon_data['id'] 
    name=pokemon_data['name']

    # to retieve the base stat of HP, attack and defence
    for key in pokemon_data['stats']: 
	    
        for value in key.items():
		    
            if key['stat']['name'] == 'hp':
                HP = key['base_stat']
			
            elif key['stat']['name'] == 'attack':
                attack = key['base_stat']
            
            elif key['stat']['name'] == 'defense':
                defense = key['base_stat']
		
        types = [t['type']['name']for t in pokemon_data['types']]
	
    pokemon_details={'id':id, 'name':name, 'HP': HP, 'attack': attack, 'defence': defense, 'types':types}
    return pokemon_details


```

```python
# get the weakness of the pokemon
def get_weakness(pokemon):
    response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon}')
    pokemon_data = response.json()
    types = [t['type']['name'] for t in pokemon_data['types']]
    weakness = []
    for t in types:
        response = requests.get(f'https://pokeapi.co/api/v2/type/{t}')
        type_data = response.json()
        damage_relations = type_data['damage_relations']
        for d in damage_relations['double_damage_from']:
            if d['name'] not in weakness:
                weakness.append(d['name'])
    return weakness

# combine all the weakness of the pokemon into a dictionary
def get_weakness_dict():
    weakness_dict = {}
    for i in range(1, 1010):
        weakness_dict[i] = get_weakness(i)
    return weakness_dict
```
