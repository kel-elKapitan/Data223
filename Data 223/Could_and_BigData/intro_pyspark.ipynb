### Initialize the entry point -SparkSession

The SparkSession object is responsible for initializing the Spark configuration and creating the necessary connections to the Spark cluster. 
It allows you to specify various configuration options:
* the number of cores to use
* memory allocation, 
* the location of input and output data

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
    .master("local[4]")\
    .appName("pyspark")\
    .getOrCreate()# Create if does not exist        

spark
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType

# define the schema for the DataFrame
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(), True),
    StructField("salary", FloatType(), True)
])

# create a list of data
data = [(1, "Alice", 25, "F", 5000.0), 
        (2, "Bob", 30, "M", 6000.0), 
        (3, "Charlie", 35, "M", 7000.0), 
        (4, "David", 40, "M", 8000.0),
        (5, "Ella", 45, "F", 9000.0),
        (6, "Frank", 50, "M", 10000.0),
        (7, "Grace", 25, "F", 5500.0),
        (8, "Henry", 30, "M", 6500.0),
        (9, "Isabella", 35, "F", 7500.0),
        (10, "Jack", 40, "M", 8500.0),
        (11, "Katie", 45, "F", 9500.0),
        (12, "Liam", 50, "M", 10500.0),
        (13, "Mia", 25, "F", 5200.0),
        (14, "Noah", 30, "M", 6200.0),
        (15, "Olivia", 35, "F", 7200.0),
        (16, "Peter", 40, "M", 8200.0),
        (17, "Quinn", 45, "F", 9200.0),
        (18, "Robert", 50, "M", 10200.0),
        (19, "Samantha", 25, "F", 5300.0),
        (20, "Tom", 30, "M", 6300.0)]

# create a DataFrame from the list of data
df = spark.createDataFrame(data, schema)

# show the DataFrame
display(df)
df.show()
```
