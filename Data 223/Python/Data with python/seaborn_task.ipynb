```python
# Import a dataset and use seaborn to visualize it
# I will pick a dataset from seaborn library and visualize it accordingly
```

```python
import pandas as pd
import seaborn as sns
```

```python

# pick a dataset from the seaborn sample datasets
financial = sns.load_dataset('dowjones')
print(financial.head())
```

```python
# the best plot to use on times series data is a line plot
sns.lineplot(x='Date', y='Price', data=financial)

```

```python
# this gives us a clear picture of the data, but it is not very pretty
# we can use the matplotlib library to make it look better
import matplotlib.pyplot as plt

# set the size of the plot
plt.figure(figsize=(16, 6))

# plot the data
sns.lineplot(x='Date', y='Price', data=financial)

# add a title
plt.title('Dow Jones Industrial Average')

# add axis labels
plt.xlabel('Date')
plt.ylabel('Price')

# rotate the x-axis labels
plt.xticks(rotation=45)


plt.show()
```
