```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# a simple scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# add a third dimension to the scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', hue_order=['Yes', 'No']) # yes, no hue order is the order the legend will follow

```

```python

# change the colour of the hue for colour blindness or corporate colours
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='winter_r') # change the colour of the hue



```

```python
hue_colours = {'Yes':'red', 'No':'black'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', palette = hue_colours)
```

```python
# custom colours from hex values
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', palette = ['#FF0000', '#000000'])
```

```python
# Count plot against a single categorical variable

sns.countplot(x='smoker', data=tips)
```

```python
# what is the distribution for the sex of the smoker
sns.countplot(x='smoker', data=tips, hue='sex')
```

```python

# relplot - create a relational plot and catplot - create a plot for categorical data
```

```python
# Relplot - default is scatter plot

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter') 
```

```python
# create a line plot with 3rd dimension
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='smoker') 
```

```python
# split the data into two plots based on the smoker column
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker') 
```

```python
# create a plot for each smoker type and split the data into two plots based on the time column
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='time', row='smoker') 
```

```python
# create a plot for each smoker type and split the data into two plots based on the time column
sns.relplot(data=tips, x='total_bill', y='tip', col='day', col_wrap=2, hue='sex', style='smoker') # col_wrap - how many columns to wrap the data into
```
